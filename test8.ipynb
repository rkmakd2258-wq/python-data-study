그래프 그리기

```python
# 모듈 불러오기
import numpy as np
# 맷플롯립은 여러 모듈로 구성됨
# 그 중 그래프를 제공하는 모듈 꺼내기
import matplotlib.pyplot as plt
# 그래프를 그릴 때 한글 깨짐 현상 발생
from matplotlib import rc
rc('font', family='Malgun Gothic')
```

```python
# 1. 선 그래프를 그리세요.
# 값 : 10부터 20까지

a = np.arange(10,21)
plt.plot(a)
```

```python
# 2. 막대 그래프를 그리세요.
# 과일 : ['사과','바나나','포도','오렌지']
# 판매량 : [50, 80, 40, 70]
a = ['사과','바나나','포도','오렌지']
b = [50, 80, 40, 70]
plt.bar(a,b)
plt. title('과일 판매')
plt. xlabel('과일')
plt. ylabel('판매량')
```

```python
# 3. 산점도 그래프를 그리세요.
x = [1, 2, 3, 4, 5]
y = [40, 55, 65, 80, 85]
plt.scatter(x, y)

plt. title('산점도 그래프')
plt. xlabel('공부 시간')
plt. ylabel('시험 점수')
```

```python
# 4. 히스토그램을 그리세요.
# - 값: 하루 유튜브 시청 시간(0~10시간 사이, 100명)
time = np.random.randint(0,11,100)
time
plt.hist(time, bins=10)
plt. title('유튜브 시청 시간 분포')
plt. xlabel('시청 시간(시간)')
plt. ylabel('사람 수')
```

여러 그래프 그리기

```python
# 하나의 도화지에 2개의 그래프(선 그래프, 막대 그래프) 그리세요.
# 데이터 : 과일 판매량
# 과일 : ['사과', '바나나', '포도', '오렌지', '복숭아']
# 판매 개수 : [50, 80, 40, 70, 60]

fig = plt.figure(figsize=(10,3))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
fruit = ['사과', '바나나', '포도', '오렌지', '복숭아']
cnt = [50, 80, 40, 70, 60]
ax1. plot(cnt)
ax2.bar(fruit, cnt)
ax1.set_title('과일 판매량')
ax2.set_title('과일 판매량')
```

```python
# 2. 하나의 도화지에 2개의 그래프(선 그래프, 막대 그래프) 그리세요.
# 데이터 : 요일별 유튜브 시청 시간
# 요일 : ['월', '화', '수', '목', '금']
# 시청 시간 : [2, 3, 4, 2, 5]

fig = plt.figure(figsize=(10,3))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
day = ['월', '화', '수', '목', '금']
time = [2, 3, 4, 2, 5]
ax1.plot(day, time)
ax2.bar(day, time)
ax1.set_title('요일별 유튜브 시청 시간')
ax2.set_title('요일별 유튜브 시청 시간')
# plt.title('요일별 유튜브 시청 시간')
```

```python
# 3. 하나의 도화지에 2개의 그래프(막대 그래프, 산점도) 그리세요.
# 데이터 : 요일별 독서 시간
# 요일 : ['월', '화', '수', '목', '금’]
# 독서 시간 : [1, 0.5, 2, 1.5, 2.5]

fig = plt.figure(figsize=(10,3))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
day = ['월', '화', '수', '목', '금']
time = [1, 0.5, 2, 1.5, 2.5]
ax1.bar(day, time)
ax2.scatter(day, time)
ax1.set_title('요일별 독서 시간 (막대 그래프)')
ax2.set_title('요일별 독서 시간 (산점도)')
```
