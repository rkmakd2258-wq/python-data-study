```python
# 1. 서울시설공단_서울도시고속도로 노선별 시간대별 교통량.csv
df = pd.read_csv('0917/서울시설공단_서울도시고속도로 노선별 시간대별 교통량.csv',encoding='euc-kr')
df
```

```python
# 노선별 총 교통량
# groupby를 하면 노선이 인덱스가 된다
result = df.groupby('노선').sum()
result['교통량']
```

```python
df_clean = df.dropna(subset=['교통량'])
result = df_clean.groupby('노선')['교통량'].mean()
result
```

```python
# 방향별 총 교통량
result = df.groupby('방향').sum()
result['교통량']
```

```python
# 시간대별 평균 교통량
result = df.groupby('시간대').sum()
result['교통량']
```

```python
# 교통량 가장 많은 시간대
df['교통량'].max()
df['교통량'].idxmax() #175(행 번호)
# 제일 교통량이 많은 행 확인
df.loc[175]['시간대']
```

```python
# 2. 서울시 버스노선별 정류장별 승하차 인원 정보.csv
df = pd.read_csv('0917/서울시 버스노선별 정류장별 승하차 인원 정보.csv',encoding='euc-kr')
df
```

```python
# 노선번호별 총 승차승객 수
result = df.groupby('노선번호').sum()
result['승차총승객수']
```

```python
# 노선번호별 총 하차승객 수
result = df.groupby('노선번호').sum()
result['하차총승객수']
```

```python
# 정류장별 총 승차승객 수
result = df.groupby('역명').sum()
result['승차총승객수']
```

```python
# 741 노선이 지나가는 정류장별 승차승객 수
df['노선번호'] == '741'
df[df['노선번호'] == '741']

df[df['노선번호'] == '741'].groupby('표준버스정류장ID').sum()
```

```python
# 승차승객 수가 100명 이상인 정류장
df['승차총승객수'] >= 100
df[df['승차총승객수'] >= 100]['역명']
```

```python
# 승차와 하차 승객 수의 합계를 "총이용객수"라는 새로운 컬럼으로 추가
df['총이용객수'] = df['승차총승객수'] + df['하차총승객수']
df
```

```python
# 3
data = pd.read_csv('0917/서울특별시 마포구_법정동별_월별_코로나19 확진자 현황.csv',encoding='euc-kr')
df = data[ ['기준연도','기준월','법정동','확진자수'] ]
df
```

```python
# 연도별 총 확진자수
df.groupby('기준연도').sum()
df.groupby('기준연도').sum()['확진자수']
```

```python
# 동별 평균 확진자수
df.groupby('법정동')['확진자수'].mean()
```

```python
# 2021년도 월별 확진자수의 합계
df['기준연도'] == 2021
df[df['기준연도'] == 2021]

df[df['기준연도'] == 2021].groupby('기준월').sum()['확진자수']
```

```python
# 확진자수가 없는 행의 개수
df['확진자수'].isnull().sum()
```

```python
# 동별 확진자수를 구하고, 큰값 순으로 정렬(5개만 출력)
df.groupby('법정동')['확진자수'].sum().head()
```

```python
# 2022년 확진자수가 가장 많은 동과 확진자수
df['기준연도'] == 2022
df[df['기준연도'] == 2022]
df[df['기준연도'] == 2022].groupby('법정동').sum()['확진자수']
# 최대값
df[df['기준연도'] == 2022].groupby('법정동').sum()['확진자수'].max()
# 최대값이 있는 행의 위치
df[df['기준연도'] == 2022].groupby('법정동').sum()['확진자수'].idxmax()
df[df['기준연도'] == 2022].groupby('법정동').sum()['확진자수']['성산동']
```
