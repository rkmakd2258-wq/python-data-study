연습문제 - 인천 마을회관 버스정류장 지도 만들기

```python
import pandas as pd
import folium 
```

```python
data = pd.read_csv('국토교통부_전국 버스정류장 위치정보_20241028.csv', encoding= 'euc-kr')
```

```python
df = data[ ['정류장명','위도','경도','도시명'] ]
```

자주 사용하는 문자열 함수  
- contains:특정 문자열이 포함됐는지
- startwith:시작하는 지
- endwith:끝나는지

```python
# 정류장 컬럼의 자료형은 object(string)
# 컬럼의 자료형이 string 이면 str 속성으로 문자열함수 사용 가능
df['정류장명'].str.contains('마을회관')
df[ df['정류장명'].str.contains('마을회관') ]
```

```python
# 1. 인천 버스정류장에서 정류장명에 ‘마을회관’이 포함된 정류장
result = df[df['도시명'] == '인천광역시']
# 2)
df1 = result[ result['정류장명'].str.contains('마을회관') ]
```

```python
# 2. 서울 버스정류장에서 정류장명에 ‘상가’가 포함된 정류장
seoul = df[df['도시명'] == '서울특별시']
seoul['정류장명'].str.contains('상가')
df2 = seoul[seoul['정류장명'].str.contains('상가')]
```

```python
# 인천정류장
df1
# 서울정류장
df2
```

필요한 데이터가 여기저기 흩어져 있음 -> 데이터 합치기  
- merge: 조건이 맞으면 연결
- concat: 그냥 이어붙이기
    - axes = 0 : 위아래로 연결. 기본값
    - axes = 1 : 옆으로 연결

```python
bus = pd.concat([df1, df2])
bus
```

지도 만들기

```python
map = folium.Map(location=[37.492642, 126.857774], zoom_start=10)
map
```

지도 위에 정류장 위치를 마커로 추가하기

```python
# 마커 올리기
# folium.Marker([37.492642, 126.857774], tooltip='정류장').add_to(map)
```

지도 위에 정류장 마커 올리기

```python
for i in bus.index:
    name = bus.loc[i, '정류장명']
    lat = bus.loc[i, '위도']
    lon = bus.loc[i, '경도']
    city = bus.loc[i, '도시명']
    # print(name, lat, lon, city)
    # 정류장을 마커로 만들기
    # 하드코딩 먼저 해서 확인하고 변수로 바꾸기
    
    # 지역에 따라 마커 색 변경
    if city == '서울특별시':
        m_color = 'blue'
    elif city == '인천광역시':
        m_color = 'red'

    
    folium.Marker([lat, lon], tooltip=name, icon=folium.Icon(color=m_color)).add_to(map)
```
