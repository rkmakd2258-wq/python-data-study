```python
# 모듈 불러오기
import numpy as np
# 맷플롯립은 여러 모듈로 구성됨
# 그 중 그래프를 제공하는 모듈 꺼내기
import matplotlib.pyplot as plt
# 그래프를 그릴 때 한글 깨짐 현상 발생
from matplotlib import rc
rc('font', family='Malgun Gothic')
```

```python
# 1. 선 그래프를 그리세요.
# 값 : 10부터 20까지

a = np.arange(10,21)
plt.plot(a)
```

```python
# 2. 막대 그래프를 그리세요.
# 과일 : ['사과','바나나','포도','오렌지']
# 판매량 : [50, 80, 40, 70]
a = ['사과','바나나','포도','오렌지']
b = [50, 80, 40, 70]
plt.bar(a,b)
plt. title('과일 판매')
plt. xlabel('과일')
plt. ylabel('판매량')
```

```python
# 4. 히스토그램을 그리세요.
# - 값: 하루 유튜브 시청 시간(0~10시간 사이, 100명)
time = np.random.randint(0,11,100)
time
plt.hist(time, bins=10)
plt. title('유튜브 시청 시간 분포')
plt. xlabel('시청 시간(시간)')
plt. ylabel('사람 수')
```

여러 그래프 그리기

```python
# 하나의 도화지에 2개의 그래프(선 그래프, 막대 그래프) 그리세요.
# 데이터 : 과일 판매량
# 과일 : ['사과', '바나나', '포도', '오렌지', '복숭아']
# 판매 개수 : [50, 80, 40, 70, 60]

fig = plt.figure(figsize=(10,3))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
fruit = ['사과', '바나나', '포도', '오렌지', '복숭아']
cnt = [50, 80, 40, 70, 60]
ax1. plot(cnt)
ax2.bar(fruit, cnt)
ax1.set_title('과일 판매량')
ax2.set_title('과일 판매량')
```

```python
# 2. 하나의 도화지에 2개의 그래프(선 그래프, 막대 그래프) 그리세요.
# 데이터 : 요일별 유튜브 시청 시간
# 요일 : ['월', '화', '수', '목', '금']
# 시청 시간 : [2, 3, 4, 2, 5]

fig = plt.figure(figsize=(10,3))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
day = ['월', '화', '수', '목', '금']
time = [2, 3, 4, 2, 5]
ax1.plot(day, time)
ax2.bar(day, time)
ax1.set_title('요일별 유튜브 시청 시간')
ax2.set_title('요일별 유튜브 시청 시간')
# plt.title('요일별 유튜브 시청 시간')
```

```python
# 3. 하나의 도화지에 2개의 그래프(막대 그래프, 산점도) 그리세요.
# 데이터 : 요일별 독서 시간
# 요일 : ['월', '화', '수', '목', '금’]
# 독서 시간 : [1, 0.5, 2, 1.5, 2.5]

fig = plt.figure(figsize=(10,3))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
day = ['월', '화', '수', '목', '금']
time = [1, 0.5, 2, 1.5, 2.5]
ax1.bar(day, time)
ax2.scatter(day, time)
ax1.set_title('요일별 독서 시간 (막대 그래프)')
ax2.set_title('요일별 독서 시간 (산점도)')
```

```python
# 모듈 불러오기
import numpy as np
import matplotlib.pyplot as plt
# 한글설정
from matplotlib import rc
rc('font', family='Malgun Gothic')
```

```python
# 1. 지점별 매출 데이터를 선 그래프로 그리세요.
x = ['서울','부산','대구','광주']
y = [150, 200, 180, 220]
plt.plot(x, y, color = 'blue', linestyle = '--',marker='o')
plt.text('서울',150, '가장 낮은 매출', color = 'red', ha='center', va='bottom')
plt.title('지점별 매출 현황')
```

```python
# 3. 요일별 운동 시간 데이터를 선 그래프로 그리세요.
x = ['월','화','수','목','금']
y = [20, 35, 25, 40, 30]

plt.plot(x, y, color='purple', linestyle = '-', marker='^')
plt.text('월',20,'최소운동',color='red',ha='center',va='bottom')
plt.title('요일별 운동 시간(분)')
```

여러 그래프 그리기

요일별 평균 기온 데이터를 그래프로 그리기

```python
# 1. 요일별 평균 기온 데이터를 DataFrame으로 만들고, 다음과 같이 그래프를 그리세요.
# 첫 번째 서브플롯: 선 그래프
# 두 번째 서브플롯: 산점도

df = pd.DataFrame({'요일':['월','화','수','목','금','토','일'],
                   '기온':[12,14,13,15,17,18,16]})
fig, axs = plt.subplots(1,2, figsize=(10,3))
sns.lineplot(data=df,x='요일',y='기온', ax=axs[0])
axs[0].set_title('요일별 기온(선)')
sns.scatterplot(data=df,x='요일',y='기온', ax=axs[1])
axs[1].set_title('요일별 기온(산점도)')
```

```python
# 2. 과일별 판매량 데이터를 DataFrame으로 만들고, 다음과 같이 그래프를 그리세요.
# 첫 번째 서브플롯: 막대 그래프
# 두 번째 서브플롯: 선 그래프
# 세 번째 서브플롯: 산점도

df = pd.DataFrame( { '과일': ['사과','바나나','포도','딸기'],
                     '판매량': [120, 80, 150, 100] } )
df
fig, axs = plt.subplots(1,3, figsize=(12,4))
sns.barplot(data=df,x='과일',y='판매량', ax=axs[0])
axs[0].set_title('과일별 판매량(막대)')
sns.lineplot(data=df,x='과일',y='판매량', ax=axs[1])
axs[1].set_title('과일별 판매량(선)')
sns.scatterplot(data=df,x='과일',y='판매량', ax=axs[2])
axs[2].set_title('과일별 판매량(산점도)')
plt.tight_layout()
```

```python
# 3. 학생 시험 점수 데이터를 DataFrame으로 만들고, 다음과 같이 그래프를 그리세요.
# 첫 번째 서브플롯: 산점도
# 두 번째 서브플롯: 막대 그래프
# 세 번째 서브플롯: 선 그래프

df = pd.DataFrame( { '공부시간': [2, 4, 6, 3, 5, 7, 8],
                     '점수': [55, 65, 75, 60, 70, 85, 90] } )
df
fig, axs = plt.subplots(3,1, figsize=(5,10))
sns.scatterplot(data=df,x='공부시간',y='점수', ax=axs[0])
axs[0].set_title('공부시간-점수(산점도)')
sns.lineplot(data=df,x='공부시간',y='점수', ax=axs[1],marker='o')
axs[1].set_title('공부시간-점수(막대)')
sns.barplot(data=df,x='공부시간',y='점수', ax=axs[2])
axs[2].set_title('공부시간-점수(선)')
plt.tight_layout()
```

```python
# 4. 과일 판매량과 학생 시험 점수 데이터를 DataFrame으로 만들고, 다음과 같이 그래프를 그리세요.
# 첫 번째 서브플롯(왼쪽 위)에는 과일 판매량 막대 그래프
# 두 번째 서브플롯(오른쪽 위)에는 과일 판매량 선 그래프
# 세 번째 서브플롯(왼쪽 아래)에는 학생 시험 점수 막대 그래프
# 네 번째 서브플롯(오른쪽 아래)에는 학생 시험 점수 산점도

df1 = pd.DataFrame( { '과일': ['사과','바나나','포도','딸기'],
                     '판매량': [120, 80, 150, 100] } )

df2 = pd.DataFrame( { '이름':['영희','철수','민수','지역'],
                      '점수':[85,90,80,100] } )

fig, axs = plt.subplots(2,2, figsize=(10,8))
# 첫번째 줄에 왼쪽 그래프:axs[0,0]
# 첫번째 줄에 오른쪽 그래프:axs[0,1]
# 두번째 줄에 왼쪽 그래프:axs[1,0]
# 두번째 줄에 오른쪽 그래프:axs[1,1]

sns.barplot(data=df1,x='과일',y='판매량', ax=axs[0,0])
axs[0,0].set_title('과일별 판매량(막대)')

sns.lineplot(data=df1,x='과일',y='판매량', ax=axs[0,1],marker='o')
axs[0,1].set_title('과일별 판매량(선)')

sns.barplot(data=df2,x='이름',y='점수', ax=axs[1,0])
axs[1,0].set_title('학생별 점수(막대)')

sns.scatterplot(data=df2,x='이름',y='점수', ax=axs[1,1])
axs[1,1].set_title('학생별 점수(산점도)')


plt.tight_layout()
```

```python
# 1. 인천 인구 데이터를 데이터프레임으로 만드세요.
df = pd.DataFrame({'구':['서구', '서구', '연수구','연수구','남동구','남동구'],
                   '인구(만명)':[45,38,35,28,55,50],
                   '동':['연희동','가좌동','송도동','옥련동','구월동','논현동']
                   })
df
```

```python
# 2. 인천 인구 데이터를 사용해서, 다음과 같이 그래프를 그리세요.
# 첫 번째 서브플롯: 막대 그래프
# 두 번째 서브플롯: 선 그래프

fig, axs = plt.subplots(1,2,figsize=(10,4))

sns.barplot(data=df,x='동',y='인구(만명)',hue='구', ax=axs[0], palette='pastel')
axs[0].text('구월동',55, '55만명', color = 'red', ha='center', va='bottom')
axs[0].set_title('인천 구동별 인구 현황 (막대)', fontsize=14, fontweight='bold')

sns.lineplot(data=df,x='동',y='인구(만명)', ax=axs[1], linestyle='--' ,marker='o')
axs[1].set_title('인천 구동별 인구 현황 (선)', fontsize=14, fontweight='bold')

plt.tight_layout()
```

```python
# 3. 상품 판매 데이터를 데이터프레임으로 만드세요.
df = pd.DataFrame({'카테고리':['식품','식품','의류','의류','전자제품','전자제품'],
                   '판매량':[120,150,80,95,200,220],
                   '상품명':['라면','과자','셔츠','신발','노트북','스마트폰'] 
                  })
df
```

```python
# 4. 상품 판매 데이터를 사용해서, 다음과 같이 그래프를 그리세요.
# 첫 번째 서브플롯: 막대 그래프
# 두 번째 서브플롯: 선 그래프

fig, axs =  plt.subplots(1,2,figsize=(10,4))
sns.barplot(data=df,x='상품명',y='판매량',hue='카테고리', ax=axs[0])
axs[0].set_title('상품별 판매량 (막대)', fontsize=14, fontweight='bold')

sns.lineplot(data=df,x='상품명',y='판매량', ax=axs[1], linestyle=':' ,marker='o')
axs[1].set_title('상품별 판매량 (선)', fontsize=14, fontweight='bold')
axs[1].text('스마트폰',220, '220개', color = 'red', ha='center', va='bottom')

plt.tight_layout()
```
