지도 만들기

```python
# 지도 객체 생성
# Map(지도의 가운데 중심, 줌레벨)
# 줌레벨이 높을 수록 지도가 확대된다
map = folium.Map(location=[37.458077,126.715780], zoom_start=10)
map
```

지도 위에 학교 위치 표시

```python
# 데이터프레임의 인덱스 확인
school.index
school.columns
# 데이터프레임 전체 데이터 확인
for i in school.index:
    name = school.loc[i, '학교명']
    typ = school.loc[i, '학교급구분']
    lat = school.loc[i, '위도']
    lon = school.loc[i, '경도']

    # 조건문으로 마커의 색 결정하기
    # 학교급구분에 따라 색 결정
    if typ == '초등학교':
        color = 'orange'
    elif typ == '중학교':
        color = 'green'
    elif typ == '고등학교':
        color = 'blue'

    # tooltip:마우스 올렸을 때만 정보가 표시됨
    # popup:클릭하면 정보가 계속 표시됨
     
    # print(i, name,typ,lat,lon)
    # 학교별 마커 만들기
    folium.Marker([lat,lon],popup=name, tooltip=name, icon=folium.Icon(color=color)).add_to(map)
```
