### 8. 맷플롯립(Matplotlib)  
데이터를 시각화하는 모듈로, 그래프와 차트를 제공

```python
# 모듈 불러오기
import numpy as np
# 맷플롯립은 여러 모듈로 구성됨
# 그 중 그래프를 제공하는 모듈 꺼내기
import matplotlib.pyplot as plt
```

```python
# 그래프를 그릴 때 한글 깨짐 현상 발생
from matplotlib import rc
rc('font', family='Malgun Gothic')
```

#### 8.1 그래프 그리기  
선그래프
- plot: 선그래프를 그래는 함수

```python
# 먼저 그래프에 담을 데이터 생성
a = np.arange(1,11)

# 숫자 배열을 그래프에 그리기
# plot(데이터)
plt.plot(a)
# x축: 우리가 다은 데이터
# y축: plt가 자동으로 설정함
```

막대 그래프
- bar: 화면에 막대 그래프를 그리는 함수

```python
# 그래프에 담을 데이터 생성
# 학급별 평균 시험 점수
a = ['1반','2반','3반','4반']
b = [80,90,85,95]

plt.bar(a,b)
plt. title('학급별 평균 시험 점수')
```

산점도
- scatter: 화면에 산점도를 그리는 함수

```python
# 산점도에 담을 데이터 생성
x = [30, 25, 10, 50, 25]
y = [9, 3, 10, 6, 2]

# x축: 첫번째 배열에 따라 자동으로 범위가 설정됨
# y축: 두번째 배열에 따라 자동으로 범위가 설정됨
# 산점도 그리기
plt.scatter(x, y)
```

히스토그램
- hist: 화면에 히스토그램을 그리는 함수

```python
# 히스토그램에 사용할 데이터 만들기
# 시험 점수 100개 만들기 (랜덤함수)
# radint: 랜덤한 정수를 ㅁ나드는 함수(시작, 끝, 개수)
scores = np.random.randint(50,101,100)
scores

# 히스토그램 그리기
# plt.hist(데이터, 구간)
plt.hist(scores, bins=5)
# 그래프 꾸미기
plt. title('시험 점수 분포')
plt. xlabel('시험 점수')
plt. ylabel('학생의 수')
```

#### 8.2 여러 그래프 그리기

figure(피겨): 흰 도화지  
- 도화지의 사이즈를 설정하거나
- 여러 그래프를 한 번에 그릴 때 사용  

axes(엑시스): 도화지 안에 들어가는 그래프

```python
# 1부터 10까지 숫자 생성
a = np.arange(1,11)
# 선그래프 그리기
plt.plot(a)
# plot함수를 쓰면 자동으로 figure(도화지)가 생성된다
```

```python
# 흰 도화지 만들기
# 가로 3인치 세로 3인치 만큼 크기 설정하기
plt.figure(figsize=(3,3))
# 도화지 위에 그래프 그리기
plt.plot(np.arange(1,11))
```

```python
# 도화지를 만들고 그 위에 그래프 여러개 그리기
# 도화지 만들기
fig = plt.figure(figsize=(10,3))

# 도화지 위에 그래프를 어떤 모양으로 그릴지 생각하기
# 두개의 그래프를 한 줄에 표시
# .add_subplot(행,열,번호)
# 1행 2열 중 1번째
ax1 = fig.add_subplot(1,2,1)
# 1행 2열 중 2번째
ax2 = fig.add_subplot(1,2,2)

# 학생 시험 점수 데이터
students = [1,2,3,4,5] #학번
score = [80,90,70,85,95] #점수

# 첫번째로 선 그래프 그리기
# 위에서 만든 엑시스 사용
ax1. plot(score)
# 두번째는 막대 그래프 그리기
ax2.bar(students, score)
```

여러 그래프 그리기2

```python
# subplots: 피겨와 엑시스를 한번에 생성하는 함수
# 위에서 작성했던 방식보다 편함
# subplots(행크기, 열크기, 캔버스사이즈)
fig, axs = plt.subplots(1,2,figsize=(10,3))
# 현재 axs는 두 개 크기의 배열
# axs[0] : 1행 2열의 첫번째 그래프
# axs[1] : 1행 2열의 두번째 그래프

# 학생 시험 점수 데이터
students = [1,2,3,4,5] #학번
score = [80,90,70,85,95] #점수
# 첫번째 그래프 그리기
axs[0].plot(score)
axs[1].bar(students, score)
```

#### 8.3 그래프 꾸미기  
색상과 선스타일

```python
# 학생 시험 성적 데이터
x = ['1반','2반','3반','4반']
y = [80, 90, 85,100]

# 선그래프
# 색상, 선스타일
# plt.plot(x, y,[옵션])
# plt.plot(x, y, color = 'red', linestyle = '-') #실선
# plt.plot(x, y, color = 'blue', linestyle = '--') #긴 점선
plt.plot(x, y, color = 'gray', linestyle = ':') #점선
```

```python
# 시험 성적 데이터
x = ['1반','2반','3반','4반']
y = [80, 90, 85,100]

# 마커 모양 꾸미기
# 원소의 위치를 표시할 때 사용
# plt.plot(x, y,[옵션])
plt.plot(x, y,marker='o') #원형 마커
# plt.plot(x, y,marker='^') #세모 마커
# plt.plot(x, y,marker='*') #별 마커
# plt.plot(x, y,marker='s') #네모 마커
```

그래프 위에 텍스트 추가하기  
- ha: 가로 (center: 가운데정렬, left, right)
- va: 세로 (top, center, bottom)
- va-bottom : 막대바의 끝에 텍스트 '아래'를 붙인다
- va-top : 막대바의 끝에 텍스트 '위'를 붙인다

```python
# 요일별 마신 커피의 수 
x = ['월','화','수','목','금','토','일']
y = [1,2,0,3,2,5,2]

# 막대 그래프 그리기
plt.bar(x, y,color = 'skyblue')

# 커피를 제일 많이 마신 날 표시
# 그래프에 텍스트 쓰기
# plt.text(위치, 문자열 문자열의 색상)
# 텍스트의 위치 조정
plt.text('토',5, '제일 많이 마신 날!', color = 'red', ha='center', va='bottom')
```

#### 8.4 시본으로 그래프 그리기

```python
# 필요한 모듈 불러오기
import pandas as pd
import seaborn as sns
# 시본도 한글 폰트 설정 필요
sns.set_theme(font='Malgun Gothic')
```

시본으로 그래프 만드는 방법
- 시본은 맷플롯립과 다르게 데이터프레임을 사용

```python
# 그래프에 넣을 시험 점수 데이터
df = pd.DataFrame( { '이름':['영희','철수','민수','지역'],
                     '점수':[85,90,80,100],
                     '반':['1반','1반','2반','2반'] } )
df
```

그래프의 종류  
- lineplot: 선 그래프
- barplot: 막대 그래프
- sctterplot: 산점도
- histplot: 히스토그램

선그래프

```python
# lineplot(데이터, x축에 사용할 열, y축에서 사용할 열)
# seaborn으로 그래프를 그릴 때는 매개변수로 값을 설정해야 한다
sns.lineplot(data=df, x='이름', y='점수')
# 그래프 제목
plt. title('학생별 점수')
```

막대 그래프 그리기

```python
# 막대 그래프 그리기
sns.barplot(data=df, x='이름', y='점수')
plt.title('학생별 점수')
```

산점도

```python
# 산점도 그리기(점 그래프)
# 산점도는 두 변수의 상관관계를 표현할 때 사용
sns.scatterplot(data=df, x='이름', y='점수')
plt.title('학생별 점수')
```

히스토그램
- 많은 양의 데이터로 구간을 나누고, 분포를 표현할 때 사용

```python
# 시험 점수 100개 생성
# 범위:50~100
# 개수:100
arr = np.random.randint(50, 101, 100)
# arr -> dataframe
df = pd.DataFrame({'점수':arr})
df
# 히스토그램 그리기
# y는 없음(데이터에 따라 자동으로 설정됨)
# bins는 구간
# 100개의 데이터를 5개의 구간으로 나눔
# 현재 y축=사람의 수
sns.histplot(data = df, x='점수',bins=5)
# y축 라벨 설정
plt.ylabel('학생의 수')
```

#### 8.5 시본으로 여러 그래프 그리기

서브 플롯:도화지 하나에 여러 그래츠를 그리는 기능

- figure(피겨):도화지(캔버스)
- axes(엑시스):그래프가 그려기는 영역(서브플록이라고도 부름)

```python
# 도화지 만들기
# subplots(행열의 크기, 캔버스의 크기)
# 피겨와 엑시스가 한번에 생성됨
fig, axs = plt.subplots(1, 2, figsize=(10,3))
# 엑시스는 행열에 따라 크기가 결정됨

df = pd.DataFrame( { '이름':['영희','철수','민수','지역'],
                     '점수':[85,90,80,100],
                     '반':['1반','1반','2반','2반'] } )
df

sns.barplot(data=df, x='이름', y='점수',ax=axs[0])
axs[0].set_title('학생 점수(막대)')

sns.lineplot(data=df, x='이름', y='점수',ax=axs[1])
axs[1].set_title('학생 점수(선)')

plt.tight_layout()
```

#### 8.6 시본으로 그래프 꾸미기

막대 그래프 꾸미기

```python
# 시험 점수 데이터
df = pd.DataFrame( { '이름':['영희','철수','민수','지연'],
                     '점수':[85,90,80,100],
                     '반':['1반','1반','2반','2반'] } )
df
```

```python
# hue: 반에 따라 막대색 칠하기
sns.barplot(data=df,x='이름',y='점수',hue='반')
```

막대 색상, 그래프 제목, xy축 라벨

```python
# 막대의 색을 파스텔 톤으로 다 다르게 칠함
sns.barplot(data=df, x='이름',y='점수', palette='pastel')
# 타이틀의 크기를 키우고 볼드체로 변경
plt.title('학생별 점수', fontsize=14, fontweight='bold')
# x축과 y축에 라벨 붙이기
plt.xlabel('학생의 이름')
plt.xlabel('학생의 점수')
```

선스타일, 마커모양, 그래프 제목, xy축 라벨

```python
# 선그래프 그리기
sns.lineplot(data=df, x='이름',y='점수', 
             linestyle=':', color='green', linewidth=3,
             marker='o')
# 그래프 제묵
plt.title('학생별 점수', fontsize=18, fontweight='bold')
# x축과 y축에 라벨
plt.xlabel('학생의 이름')
plt.ylabel('학생의 점수')
```
